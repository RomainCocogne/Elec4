## Eliminate Beat Frequency

In the previous computer exercice, you were asked to eliminate beat frequency occuring when you have two close frequencies (here $f_1 = 440 Hz$ and $f_2 = 442 Hz$.).

Based on the conclusions drawn when answering the previous question, select an IIR filter and design a filter eliminating the 442 Hz frequency.  What is the ratio between the computational complexity of the FIR filter and the IIR filter ?

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy.signal as signal

def plot_template(Fr, Att,Fs,axes,togiir,title):
    '''
    plot_template(frequencies, attenuations, Fs)
    frequencies holds an array with the corner frequencies.  It has to start
     with 0 and terminate with Fs/2
        example : low pass Fr=np.array([0,100,110,Fs/2])

    attenuation holds an array with the corresponding attenuations
        example : low pass Att=np.array([2,-60])
                (ripple of 2 dB, Attenuation of -60 dB)
    Fs holds the sampling frequency
    axes [minx (in Hz), maxx (in Hz), min y (in dB), max y (in dB)]
            Very useful to zoom on parts of the frequency response
    togiir = 1 if you design an IIR filter (in the IIR filter, the 
                bandpass maximum gain is 0dB, so the amplitude is between
                0 and 0-ripple in dB.  For FIR filters, the mean
                attenuation is 0 dB)
    title : title of the plot
    
    Committed by L. Deneire, Nov. 20 2018
    '''
    min_y=axes[2]
    togiir=-togiir
    p=[]
        # a polygon inside which the frequency response has to live !
    tog=(Fr.size>4)
        # for internal purpose
    ftype=int(Att[0]>0)
        #ftype : 1 if it starts with low attenuation (e.g. low pass)
        #      : 0 if it starts with high attenuation (e.g. high pass)
    if (ftype==1):
        for i in np.arange(0,Fr.size,2):
           if (np.mod(i,4)==0):
                p=p+[[Fr[i], Att[i//2]/2*togiir-Att[i//2]/2],[Fr[i+1], Att[i//2]/2*togiir-Att[i//2]/2],[Fr[i+1], min_y]]
           else:
                p=p+[[Fr[i-1], min_y],[Fr[min(i+1+int(tog),Fr.size-1)], min_y]]
        for i in np.arange(Fr.size-2,-1,-2):
           if (np.mod(i,4)==0):
                p=p+[[Fr[i+1], Att[i//2]/2+Att[i//2]/2*togiir],[Fr[max(0,i-1)], Att[i//2]/2+Att[i//2]/2*togiir]]
           else:     
                p=p+[[Fr[i+1], Att[i//2]],[Fr[i], Att[i//2]],[Fr[i], Att[i//2-1]/2+Att[i//2-1]/2*togiir]]
    else:
        for i in np.arange(0,Fr.size,2):
           if (np.mod(i,4)==0):
                p=p+[[Fr[i], Att[i//2]],[Fr[i+1], Att[i//2]],[Fr[i+1], Att[i//2]/2]]
           else:
                p=p+[[Fr[i-1], Att[i//2]/2+Att[i//2]/2*togiir],[Fr[min(i+1+int(tog),Fr.size-1)], Att[i//2]/2+Att[i//2]/2*togiir]]
        for i in np.arange(Fr.size-2,-1,-2):
           if (np.mod(i,4)==0):
                p=p+[[Fr[i+1], min_y],[Fr[max(0,i-1)], min_y]]
           else:     
                p=p+[[Fr[i+1], Att[i//2]/2*togiir-Att[i//2]/2],[Fr[i], Att[i//2]/2*togiir-Att[i//2]/2],[Fr[i], min_y]]
    polygon = mpatches.Polygon(p, closed=True, facecolor='r', alpha=0.3)
    plt.gca().add_patch(polygon)
    plt.title(title)
    plt.xlabel(r'$f$')
    plt.ylabel(r'$|H(f)|$ in dB')
    plt.axis(axes)
    plt.grid();

    
 
```

##  Example - compare FIR rectangular window and optimized FIR

```python

Fs=1000;  

Fr4=np.array([0,100,110,Fs/2])
Att4=np.array([2,-60])



# Estimate the number of taps based on Kaiser's window
numtaps, beta = signal.kaiserord(60, 10/(0.5*Fs))

# Design a FIR filter with a rectangular window
h1 = signal.firwin(numtaps, 104, window='rectangular',scale=False, nyq=0.5*Fs)
hb = signal.firwin(numtaps, 104, window='hamming',scale=False, nyq=0.5*Fs)

# Design an optimal FIR filter based on the Remez Algorithm
lpf = signal.remez(numtaps, Fr4/Fs, np.array([1.0, 0.0]))

w, Hk = signal.freqz((lpf))
w, Hr = signal.freqz(h1)
w, Hb = signal.freqz(hb)

axes=[0, Fs/2, -120, 3]
plt.figure(figsize = (15, 10))
plot_template(Fr4, Att4, Fs, axes,0,'Low Pass FIR filter comparison rectangular window vs optimized')
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hr)))
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hk)))
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hb)))

plt.legend(['rectangular','Remez','hamming'])
```

## example : comparison zoomed at the corner

```python
axes=[75, 200, -80, -40]
plt.figure(figsize = (15, 10))
plot_template(Fr4, Att4, Fs, axes,0,'Low Pass FIR filter comparison rectangular window vs optimized')
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hr)))
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hk)))
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hb)))

plt.legend(['rectangular','Remez','hamming'])
```

## Example : chebyshev-II IIR filter vs optimized FIR

```python
wp = 0.2
ws = 0.22
gpass=Att4[0]
gstop=-Att4[1]
b,a=(signal.iirdesign(wp, ws, gpass, gstop, ftype='ellip', output='ba'))
w, Hiir = signal.freqz(b,a)


axes=[0, Fs/2, -100, 3]
plt.figure(figsize = (15, 10))
plot_template(Fr4, Att4, Fs, axes,1,'Low Pass Filter comparison FIR vs IIR')
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hk)))
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(Hiir)))
plt.legend(['FIR','IIR'])
```

```python
## example of beat frequency at 440 and 442 Hz
```

```python
import IPython

Fs=8000

f1=440
f2=442
om1=2*np.pi*f1
om2=2*np.pi*f2
t=np.arange(0,5,1/Fs)
y=np.sin(om1*t)+np.sin(om2*t)
IPython.display.Audio(y, rate=Fs)
```

## Spectrogram 

```python

f, P_spec = signal.periodogram(y, Fs, scaling='spectrum')

plt.semilogy(f, P_spec)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')
plt.grid()
plt.show()
#  Zoom between 430 and 450 Hz
fc=np.where(np.logical_and(( f>430) , (f<450)))
plt.semilogy(f[fc],P_spec[fc])
```

## Filtering with FIR filter - estimating the time taken in milliseconds

```python


import timeit

width=f2-f1             # width in Hertz
                    # determine the order of the filter for 20 dB attenuation
numtaps, beta = signal.kaiserord(20, width/(0.5*Fs))
print('Number of taps :'+str(numtaps)+'\n')

                    # design the optimal FIR filter
lpf = signal.remez(numtaps, np.array([0, f1, f2-0.2, Fs/2])/Fs, [1.0, 0.0])
                    #IIR

b,a=(signal.iirdesign(2*f1/Fs, 2*f2/Fs, 2, 20, ftype='ellip', output='ba'))

# plot impulse response
                    # compute the frequency response in amplitude
w, h = signal.freqz(lpf,1,512*32)  # take more than 512 points in frequency to enhance resolution
wiir, hiir = signal.freqz(b,a,512*32)

                    # perform the filtering
zi = signal.lfilter_zi(lpf, 1)
z,_=signal.lfilter(lpf, 1, y, zi=zi*y[0])

ziiir = signal.lfilter_zi(b, a)
ziir,_=signal.lfilter(b, a, y, zi=ziiir*y[0])


                    # estimate the time taken to filter (platform dependent of course)
mytime_lfilter=timeit.timeit(lambda: signal.lfilter(lpf, 1, y, zi=zi*y[0]),number=100)*10
mytime_convolve=timeit.timeit(lambda: signal.convolve(lpf,y),number=100)*10
mytime_iirlfilter=timeit.timeit(lambda: signal.lfilter(b, a, y, zi=ziiir*y[0]),number=100)*10
print('Time to perform filtering with lfilter= '+str(mytime_lfilter)+' milli-seconds \n')
print('Time to perform filtering with convolve= '+str(mytime_convolve)+' milli-seconds \n')
print('Time to perform filtering with iir= '+str(mytime_iirlfilter)+' milli-seconds \n')


#  Plot template and frequency response of the filter

plt.figure() 
axes=[400,500,-80,5]                 # to zoom on region of interest
plt.figure(figsize = (15, 10))
Fr=np.array([0,440,442,Fs/2])        # template
Att=np.array([2,-20])                # somewhat arbitrary
plot_template(Fr, Att,Fs,axes,0,'Beat elimination by IIR')

# Compute and plot the input and output power spectral density

f, P_spec = signal.periodogram(z, Fs, scaling='spectrum')
fy, P_specy = signal.periodogram(y, Fs, scaling='spectrum')


plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(h)))
plt.plot(fy, 10*np.log10(P_specy))
plt.plot(f, 10*np.log10(P_spec))
plt.plot(w/(2*np.pi)*Fs, 20*np.log10(np.abs(hiir)))
plt.legend(['Frequency response','input power spectrum','output power spectrum'])


# You can hear the beat has (almost) disappeared

IPython.display.Audio(ziir, rate=Fs)
```
